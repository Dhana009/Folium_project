## add a marker to the map

```python
import folium as f

maps = f.Map(location=[46.961580, -102.560670],tiles="OpenStreetMap",
            zoom_start=4)

f.Marker(location=[47.606209, -122.332069],popup='Seattle',icon = f.Icon(icon='cloud')).add_to(maps)
f.Marker(location=[37.774929, -122.419418],popup='San Fransisco',icon=folium.Icon(color='green')).add_to(maps)
f.Marker(location=[41.878113, -87.629799],popup='Chicago',icon=folium.Icon(color='red', icon='envelope')).add_to(maps)

maps.save('map.html')
```

## add a custom marker to the map

```python
import folium as f

maps = f.Map(location=[46.961580, -102.560670],tiles="OpenStreetMap",
            zoom_start=4)

f.Marker(location=[47.606209, -122.332069],popup='Seattle',icon = f.Icon(icon='cloud')).add_to(maps)
f.Marker(location=[37.774929, -122.419418],popup='San Fransisco',icon=f.Icon(color='green')).add_to(maps)
f.Marker(location=[41.878113, -87.629799],tooltip='Chicago',popup='Chicago',icon=f.Icon(color='red', icon='envelope')).add_to(maps)

#creating custom icons
spidermanicon = f.features.CustomIcon('spiderman.png',icon_size=(100,100))
hulkicon = f.features.CustomIcon('hulk.png',icon_size=(100,100))

#descriptions about heros
spidermanpop = "<strong>spiderman</strong><br>realname : dj<br>city of birth : tpt"
hulkpop = "<strong>hulk</strong><br>realname : hulk dj<br>city of birth : tpt"

f.Marker(location=[40.743720, -73.822030],popup=spidermanpop,tooltip='spiderman',icon=spidermanicon,draggable=True).add_to(maps)
maps.save('map.html')
```

## create a route in a map

```python
import folium as f
import os
import base64

maps = f.Map(location=[13.652563, 79.462942],zoom_start=15)

f.GeoJson('walk.json', name='walk').add_to(maps)

tooltip = 'click to see picture'
html = '<img src="data:image/png;base64,{}">'.format

picture1 = base64.b64encode(open('spiderman.png','rb').read()).decode()
iframe1 = f.IFrame(html(picture1), width=200 , height=200)
popup1 = f.Popup(iframe1,max_width=220)
icon1 = f.Icon(color='red', icon='glyphicon-home')

marker1 = f.Marker(location=[13.665399330943416,79.46139703007901],
                   tooltip=tooltip,popup=popup1,icon=icon1).add_to(maps)
            

maps.save('walk.html')
```
